```python
from pycocotools.coco import COCO
import numpy as np
```

```python
dataDir='coco'
dataType='train2017'
annFile='{}/annotations/instances_{}.json'.format(dataDir,dataType)
```

```python
# initialize COCO api for instance annotations
coco=COCO(annFile)
```

```python
images_per_category=10
```

```python
# display COCO categories and supercategories
cats = coco.loadCats(coco.getCatIds())
categoryNames=[cat['name'] for cat in cats]

imgIds=[]

for categoryName in categoryNames:
    currentCatIds = coco.getCatIds(catNms=[categoryName])
    currentImgIds = coco.getImgIds(catIds=currentCatIds)

    # Shuffle imgIds to get a random subset for each category
    np.random.shuffle(currentImgIds)
    # Select first N images add to imgIds
    imgIds += currentImgIds[:images_per_category]
    
coco.download(tarDir='{}/download/{}'.format(dataDir, dataType), imgIds=imgIds)
```

```python
imgs = coco.loadImgs(imgIds)

with open('dataset.txt', 'w') as file:
    for img in imgs:
        file.write(f"./{dataDir}/{dataType}/{img['file_name']} {img['id']}\n")
```
